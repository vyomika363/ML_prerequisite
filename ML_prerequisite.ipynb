#Question1

We have 2 matrices A and B, each with n*k and k*m
So while multiplying, k elements of A and multiplied with k elements of B, and then they are added, so per each element of the final matrix of A*B, we have k multiplications and (k-1) additions
The final matrix has m*n elements
SO total number of mutiplications and additions are mnk and mn(k-1) respectively.


#Question 2


```python
import numpy as np
import random
import time

m=random.randint(1,100)
n=random.randint(1,100)
k=random.randint(1,100)
#A is n*k, B is k*m
A=[[random.random() for _ in range(k) ] for _ in range(n)]
B=[[random.random() for _ in range(m) ] for _ in range(k)]

def element(A,B):
  ele=[[0 for _ in range(len(B[0]))] for _ in range(len(A))]
  for i in range(len(A)):
    for j in range(len(B[0])):
      for k in range(len(B)):
        ele[i][j]+=A[i][k]*B[k][j]
  return ele


start1=time.time()
multip=element(A,B)
end1=time.time()
print(f"LIST method:  {end1-start1:.4f}")

A_numpy=np.array(A)
B_numpy=np.array(B)
start2=time.time()
multip_numpy=np.dot(A_numpy,B_numpy)
end2=time.time()
print(f"NUMPY method:  {end2-start2:.4f}")
```

As seen by the result, numpy method takes the least time.

Numpy method is faster due to the lack of loops

#Question 3

To find the highest element, we will need to scan through all the elements of the array, so time complexity for highest is n.

To find secong highest, we will need to scan twice, first n, and then n-1
Median is the n/2 th position element.

Therefore time complexity is n+ n-1 + n-2 + ... + (n-(n/2))

Therefore, the timecomplexity is O(n^2).

We can try sorting method, then find the median index value, sorting uses quicksort, which has a Timecomplexity O(nlogn)

```python
n=random.randint(1,10000)
A=[random.random() for _ in range(n)]

#method1
def brute(A):
  A_copy = A[:]
  ans=[]
  for _ in range(len(A)):
    x=A_copy[0]
    for i in range(len(A_copy)):
      if x< A_copy[i]:
        x=A_copy[i]
    ans.append(x)
    A_copy.remove(x)
  if len(A) % 2 == 1:
    return x
  else:
    y= A_copy[0]
    for i in range(len(A_copy)):
      if y < A_copy[i]:
        y = A_copy[i]
    return (x + y) / 2

start1=time.time()
met1=brute(A)
end1=time.time()
print(f"Brute force method:  {end1-start1:.4f}")

#using sorting
def sort(A):
  A_copy=A[:]
  A_copy.sort()
  if len(A) % 2 == 1:
    return A_copy[len(A) // 2]
  else:
    return (A_copy[len(A) // 2] + A_copy[len(A) // 2 - 1])/2

start2=time.time()
met2=sort(A)
end2=time.time()
print(f"Sorting method:  {end2-start2:.4f}")

#using np.median()

start3=time.time()
met3=np.median(A)
end3=time.time()
print(f"Numpy method:  {end3-start3:.4f}")
```
